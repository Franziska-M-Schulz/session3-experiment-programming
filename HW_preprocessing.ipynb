```python
import pandas as pd
import numpy as np
```

```python
participant_1 = pd.read_csv('output_1.csv', sep = ';')
participant_2 = pd.read_csv('output_2.csv', sep = ';')
participant_3 = pd.read_csv('output_3.csv', sep = ';')
participant_4 = pd.read_csv('output_4.csv', sep = ';')

participants = [participant_1, participant_2, participant_3, participant_4]

participants = pd.concat(participants)
participants.to_csv('participants.csv', index=False)

participants_merged = pd.read_csv('participants.csv')

#participants_merged
```

```python
conditions = [[x[0] in x[1] for x in zip(participants_merged['name'], participants_merged['image'])],[x[0] not in x[1] for x in zip(participants_merged['name'], participants_merged['image'])]]

choices = ["match", "no_match"]

participants_merged["exp_condition"] = np.nan
participants_merged["exp_condition"] = np.select(conditions, choices)

#participants_merged
```

```python
accuracy = []

for index, row in participants_merged.iterrows():
    if row ['key'] == 'y' and row['exp_condition'] == 'match':
        accuracy.append(1)
    elif row ['key'] == 'n' and row['exp_condition'] == 'no_match':
        accuracy.append(1)
    else:
        accuracy.append(0)

participants_merged["accuracy"] = np.array(accuracy)
#participants_merged

#Usually, I would remove the incorrect answers

participants_merged.to_csv('participants_preprocessed.csv', index=False)
```
